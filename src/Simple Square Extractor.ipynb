```python
import cv2
import numpy as np
from util import listdir_nohidden
import matplotlib.pyplot as plt

board_dir = "./Boards/"

board_filenames = listdir_nohidden(board_dir)

board_imgs = [cv2.imread(board_dir+f, 0) for f in board_filenames]

board = board_imgs[9]
plt.imshow(board, cmap="gray")
plt.axis("off")
plt.show()
```

```python

def extract_squares(board):
    squares = []
    ww, hh = board.shape
    w = int(ww / 8)
    h = int(hh / 8)
    for i in range(8):
        for j in range(8):
            squares.append(board[i*w:(i+1)*w, j*h:(j+1)*h])
    return squares

squares = extract_squares(board)

plt.imshow(squares[7], cmap="gray")
plt.axis("off")
plt.show()
```
